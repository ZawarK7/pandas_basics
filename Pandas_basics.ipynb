```python
import pandas as pd
import numpy as np
```

```python
# Create dataset with name and age columns
data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave', 'Emily', 'Frank', 'Grace', 'Henry', 'Isabella', 'Jack', 'Kate', 'Liam', 'Mia', 'Noah', 'Olivia', 'Peter', 'Quinn', 'Rachel', 'Sophia', 'Tom'],
        'age': [22, 30, 45, 27, 31, 29, 40, 25, 28, 29, 31, 21, 23, 35, 26, 27, 33, 37, 28, 26]}
df = pd.DataFrame(data)

# Filter out records where age is equal or less than 30
df_filtered = df[df['age'] <= 30]

# Create dataset with name and gender columns
data2 = {'name': ['Alice', 'Bob', 'Charlie', 'Dave', 'Emily', 'Frank', 'Grace', 'Henry', 'Isabella', 'Jack', 'Kate', 'Liam', 'Mia', 'Noah', 'Olivia', 'Peter', 'Quinn', 'Rachel', 'Sophia', 'Tom'],
         'gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'F', 'F', 'M']}
df2 = pd.DataFrame(data2)

# Merge the two datasets
merged_df = pd.merge(df_filtered, df2, on='name')

print(merged_df)
print("\n")
print(df_filtered)
```

```python
merged_df.to_csv('merged_df.csv', index=False)
```
